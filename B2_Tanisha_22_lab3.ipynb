```python
def reconstruct_path(parent, start, goal):
    if goal not in parent and goal != start:
        return None
    path = []
    current = goal

    while current != start:
        if current not in parent:
            return None
        path.append(current)
        current = parent[current]
    path.append(start)

    path.reverse()
    return path
path = reconstruct_path(parent, start, goal)

print("\nShortest path from", start, "to", goal, ":")
print(path)
```
